```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("data_csv.csv")
plt.figure(figsize=(10,7))
x=data["Value"]
plt.hist(x,bins=20,color="green")
plt.title("Value")
plt.xlabel("Value")
plt.ylabel("count")
plt.show
```
